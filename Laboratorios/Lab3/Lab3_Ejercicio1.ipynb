```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
A = 5
phi = 0
f = 100
w = 2*np.pi*f

fs = 50*f
Ts = 1/fs
n = np.arange(1000)
t = n*Ts

x = A*np.sin(w*t+phi)
```

```python
plt.plot(t,x)
plt.xlim(0,0.1)
```

```python
plt.stem(n,x)
plt.xlim(0,100)
```

```python
plt.magnitude_spectrum(x, Fs=fs, scale='dB')
plt.grid(ls=":")
plt.xlim(0,400)
plt.show()
```
